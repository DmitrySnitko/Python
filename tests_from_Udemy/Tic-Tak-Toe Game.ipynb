```python
from IPython.display import clear_output

def board_print(board):
    # Function to print current board
    
    clear_output()
    
    for _ in range(1,10,3):
        line = ['',board[_],'|',board[_+1],'|',board[_+2],'']
        print(" ".join(line))
        if _<7: print('---|---|---')
            
def player_choise():
    # Function to choose who will play as X and who as O
    
    player1 = ''
    
    while player1 != 'X' and player1 != 'O':
        player1 = input('Player1, do you want to be "X" or "O"? ')
    
    if player1 == 'X':
        print('Player1 will start the game.')
        player2 = 'O'
        marker_x = 'Player1'
        marker_o = 'Player2'
    else:
        print('Player2 will start the game.')
        player2 = 'X'
        marker_x = 'Player2'
        marker_o = 'Player1'
        
    return(player1, player2, marker_x, marker_o)    

def game_start(player_x):
    # Request for the game start. player_x is the one who plays for X and starts the game.
    answer = ''
    while answer != 'Yes' and answer != 'No':
        answer = input("Let's start! {}, are you ready? Print Yes or No: ".format(player_x))
    return answer
    
def new_game():
    # Request whether to play more   
    answer = ''
    
    while answer != 'Yes' and answer != 'No':
        answer = input("Do you want to play again? Print Yes or No: ")
        clear_output()
    
    if answer == 'Yes':
        return True
    else:
        return False

def check_win(my_board, marker):
    #my_board = board
    result = False
    index_text = ''
    
    # Get all markes positions as text string
    for _ in range(0,len(my_board)):
        if my_board[_] == marker:
            index_text += str(_)
        
    win_combinations = ['123', '456', '789', '147', '258', '369', '159', '357']
    
    # Check if any combination is found
    for _ in win_combinations:
        if _ in index_text:
            result = True
            break
    
    return result
```

```python
while True:
    
    # Welcom message
    print('Welcom to Tic-Tak-To game!')

    # Making a player choise and writing it to variables
    player1_marker, player2_marker, player_x, player_o = player_choise()

    # The game itself
    player_answer = game_start(player_x)
    
    # Making an empty board for the start
    board = [' ']*10

    if player_answer == 'No':
        print("OK. Let's play another time. Bye!")
        break
    else:
        # Clear screen, print current board and ask for next move
        board_print(board)
        
        for _ in range(1,9,2):
            
            # Choosing position for X
            position = int(input('{}, choose your next position for X (1-9): '.format(player_x)))
            while (board[position] != ' ') or (position not in range(1,10)):
                #board_print(board)
                print('This position not correct or occupied, choose another one between 1 and 9!')
                position = int(input('{}, choose your next position for X (1-9): '.format(player_x)))
            board[position] = 'X'
            board_print(board)
            
            # Check if X has won the game and whether it wants to play more.
            if _ >= 5 and check_win(board, 'X'):
                print('Congratulations to {}! You have won the game!'.format(player_x))
                break

                  
            # Choosing position for O
            position = int(input('{}, choose your next position for O (1-9): '.format(player_o)))
            while (board[position] != ' ') or (position not in range(1,10)):
                #board_print(board)
                print('This position not correct or occupied, choose another one between 1 and 9!')
                position = int(input('{}, choose your next position for O (1-9): '.format(player_o)))
            board[position] = 'O'
            board_print(board)
            
            # Check if O has won the game and whether it wants to play more.
            if _ >= 5 and check_win(board, 'O'):
                print('Congratulations to {}! You have won the game!'.format(player_o))
                break
        else:
            print('Noone has won this game.')
            
        # Ask for a new game
        
        if not new_game():
            break
```
